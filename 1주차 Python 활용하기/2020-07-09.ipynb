## Class

```python
# 딕셔너리 : {}
class Student:
    def __init__(self, name, korean, math, english, science):
        self.name=name
        self.korean=korean
        self.math=math
        self.english=english
        self.science=science
        
    def __str__(self):
        return self.name + str(self.korean) + str(self.math) + str(self.english) + str(self.science)
        
    def get_name(self): # 이름을 변경할때는 set_name
        return self.name 
    
    def set_name(self, name): 
        self.name=name
        
    def get_sum(self):
        return self.korean + self.math + self.english + self.science
    
    def get_avg(self):
        return (self.korean + self.math + self.english + self.science)/4
    
    def get_min(self):
        return min( str(self.korean) + str(self.math) + str(self.english) + str(self.science))
    
a= Student("가나다" ,45 ,55 ,65 ,75)
a.set_name("마다라")
print(a)
print(a.get_sum())
print(a.get_avg())
print(a.get_min())
print(type(a.get_sum))
```

```python
# 이미지 받아오기
from selenium import webdriver
import urllib.request
import time
from selenium.webdriver.common.keys import Keys

options = webdriver.ChromeOptions()
#options.add_argument('headless') #화면 표시 X
options.add_argument("disable-gpu")   
options.add_argument("lang=ko_KR")    
options.add_argument('user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36')  # user-agent 

driver = webdriver.Chrome(executable_path='./chromedriver', options=options)
driver.get('https://naver.com')

driver.find_element_by_xpath('//*[@id="query"]').send_keys('바나나') # 검색창에 입력
driver.find_element_by_xpath('//*[@id="query"]').send_keys(Keys.ENTER)

#이미지 클릭
driver.find_element_by_xpath('//*[@id="lnb"]/div/div[1]/ul/li[2]/a/span').click()

for i in range(1, 6, 1): # 1부터 5까지 반복
    time.sleep(1) #1초 delay
    try: 
        #첫번째 방법으로 실행해서 오류가 발생하면 except로 이동
        img = driver.find_element_by_xpath('//*[@id="_sau_imageTab"]/div[1]/div[2]/div['+ str(i) +']/a[1]/img')
    except Exception as e:
        #첫번째 방법으로 실패한 경우에 2번째 시도
        img = driver.find_element_by_xpath('//*[@id="_sau_imageTab"]/div[2]/div[2]/div['+ str(i) +']/a[1]/img')
    
    #src태그의 이미지 주소를 가져옴.    
    file = img.get_attribute("src")
    
    #이미지 주소를 이용하여 파일 다운로드
    urllib.request.urlretrieve(file, './banana'+str(i)+".jpg") # naver1.jpg, naver2.jpg 순으로 저장

# 크롬 닫기
driver.close() 
```
