```python
# 최대값, 최소값 구하기
a=[277,155,300,256]

max=a[0] #초기값을 선정
min=a[0]
for i in a: # 라스트 값을 한번씩 다 써봄
    if max<i: # 초기값 보다 큰 수가 나오면 i값에 대입
        max=i
        
    if min>i: # 초기값보다 작은 수가 나오면 i값에 대입
        min=i
print("최대값은={}".format(max))
print("최소값은={}".format(min))
```

```python
# 순위 구하기
a=[4,65,57,1,5]
b=[1,1,1,1,1]

for i in range(len(a)): # 리스트의 길이 0부터 5보다 작을때까지
    for j in range(len(a)):
        if a[i]<a[j]: # 비교했을때
            b[i]=b[i]+1 # i가 더 작을 경우 해당 하는 b 리스트에 값 1씩 증가
print(a)
print(b)
```

```python
import pymongo

conn = pymongo.MongoClient('192.168.99.100',32767) #docker 서버 접속
#conn = pymongo.MongoClient('localhost',27017)      #윈도우
db = conn.get_database("db1") # db1생성
table = db.get_collection("member") #collection 생성

def mymenu():
    print("0. 종료")
    print("1. 추가")
    print("2. 삭제")
    print("3. 수정")
    print("4. 목록")
    n = int(input("숫자를 입력하세요?"))
    return n

def myinsert():
    na = input("이름 입력?")
    te = input("연락처 입력?")
    ag = int(input("나이 입력?"))
    em = input("이메일 입력?")
    table.insert_one({"name":na, "tel":te, "age":ag, "email":em})

def mydelete():
    na = input("삭제할 이름 입력?")
    table.delete_one({"name":na})

def myupdate():
    na = input("수정할 이름 입력(변경불가)?")
    te = input("연락처 입력?")
    ag = int(input("나이 입력?"))
    em = input("이메일 입력?")
    table.update_one({ "name" : na }, { "$set" : { "tel" : te, "age": ag, "email":em } })
def myselect():
    data1 = table.find({},{'_id':False}) #가져오기
    for i in data1:
        print(i)

        
while True: #반복문 조건 True => 무한반복문
    n = mymenu()
    
    if n == 0:
        break #반복문 종료
        
    if n == 1: #추가
        myinsert()
    
    if n == 2: #삭제
        mydelete()
    
    if n == 3: #수정
        myupdate()
        
    if n == 4: #목록
        myselect()
```

```python
# 함수 만들기
def mysum(a,b):
    return a+b

def mymin(x): #리스트로 값이 넘어옴
    min=x[0]
    for i in x:
        if min>i:
            min=i
    return min
    
    
# 만든 함수 호출하기
print(mysum(4,5))
print(mymin([1,2,0,4,5])) # x=[1,2,0,4,5]
```
