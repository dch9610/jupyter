# 5.DataFrame 열(Column) 컬럼

```python
# Age_0 열을 데이터프레임에 생성 후에 0의 값을 채우기
tatinc["Age_0"] = 0
tatinc.head()
```

```python
# Age_by_10 열을 생성하고 Age*10 값을 채우기
tatinc["Age_by_10"]=tatinc["Age"]*10
tatinc.head()
```

```python
# Family_No 열을 생성하고 탑승객 남성에 여성과 아이의 수를 합쳐서 가족으로 표현
tatinc["Family_No"]=tatinc["SibSp"] + tatinc["Parch"]+1
tatinc
```

```python
# 생성한 Age_by_10에 100을 더해보면
tatinc["Age_by_10"]=tatinc["Age_by_10"]+100
tatinc.head()
```

# DataFrame에서 데이터 삭제

```python
# 데이터프레임에서 drop 함수로 Age_0을 삭제, 이 때 axis=1 인자를 넣어 열을 삭제한다.
tatinc_drop=tatinc.drop('Age_0',axis=1)
tatinc_drop.head()
```

```python
# drop_result에 복수의 열을 삭제하며, inplace=True 인자는 해당 drop한 열을 반환하지 않고, 기존의 데이터프레임을 생성하는 것
drop_result=tatinc.drop(["Age_by_10","Family_No"],axis=1,inplace=True)
print('inplace=True로 drop 후 반환하는 값: ',drop_result)
tatinc.head()
```

```python
# 행을 삭제시 행 index를 지정
tatinc.drop([0,1,2,5], axis=0,inplace=True)
tatinc.head()
```

```python
# 열을 시리즈 반환
series_fair=tatinc['Fare']
print('Fair series Max',series_fair.max())
print('Fair Series Min:',series_fair.min())
print('Fair Series Sum: ',series_fair.sum())
print('Fair Serise +3; \3',(series_fair +3).head())
```
